```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d 

```

```python
x1 = np.linspace(0, 10, 150)
x2 = np.linspace(0, 35, 150)
x3 = np.linspace(0, 20, 150)
x4 = np.linspace(0, 25, 150)
x5 = np.linspace(0, 30, 150)
x6 = np.linspace(0, 40, 150)
x7 = np.linspace(0, 50, 150)
x8 = np.linspace(0, 450, 150)

x = np.column_stack((x1, x2, x3, x4, x5, x6,x7,x8)) 
x.shape
print(x)
```

```python
f = lambda x: 0.5 * np.exp(-0.5*x)
y = f(x).sum(axis=1)
y.shape
```

```python
noise = np.random.rand(150) / 5
y = y + noise
print(y)
```

```python
ax = plt.axes(projection ='3d')
ax.plot3D(x1, x2, y)
plt.show()
```

```python
import pandas as pd
data = {'X': x, 'y': y}
df = pd.DataFrame(data)
df.to_csv('synthetic_data_exp.csv', index=True)
```
