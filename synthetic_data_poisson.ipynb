```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
X = np.linspace(1,19,150, dtype = np.int64)
print(X)
```

```python
fact = np.vectorize(math.factorial)

f = lambda X: (8**X*np.exp(-8))/(X*fact(X))
y = f(X)
print(y)
```

```python
noise = np.random.rand(150) /1000
y = y + noise
```

```python
plt.plot(X,y)
plt.show()
```

```python
import pandas as pd

data = {'X': X, 'y': y}
df = pd.DataFrame(data)
    
df.to_csv('synthetic_data_poisson.csv', index=True)
```
