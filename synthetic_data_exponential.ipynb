```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,150)
print (x)
```

```python
f=lambda x: 0.5*np.exp(-0.5*x)
y=f(x)
print(y)
```

```python
noise = np.random.rand(150) / 10
y = y + noise
```

```python
plt.plot(x, y)
plt.show()
```

```python
import pandas as pd

data = {'X': x, 'y': y}
df = pd.DataFrame(data)

df.to_csv('synthetic_data_exp.csv', index=True)
```
